## Research Agent

A simple **agentic workflow** designed to simulate reflective thinking in a writing task.

* **Step 1 – Drafting:** Call the LLM to generate an initial draft of an essay based on a simple prompt.
* **Step 2 – Reflection:** Reflect on the draft using a reasoning step.
* **Step 3 – Revision:** Apply the feedback from the reflection to generate a revised version of the essay.

```python
from dotenv import load_dotenv
load_dotenv()
import aisuite as ai
# Define the client.
CLIENT = ai.Client()
```

```python
import unittests
```

* **Step 1 – Drafting:**

```python
def generate_draft(topic: str, model: str = "openai:gpt-4o") -> str:

    prompt = f"Write a professional and informative initial draft for an essay about: {topic}."

    response = CLIENT.chat.completions.create(
        model=model,
        messages=[{"role": "user", "content": prompt}],
        temperature=1.0,
    )

    return response.choices[0].message.content
```

```python
#test
unittests.test_generate_draft(generate_draft)
```

* **Step 2 – Reflection:**

```python

def reflect_on_draft(draft: str, model: str = "openai:o4-mini") -> str:
    prompt = f"""
    You are an expert editor. Critically evaluate the following essay draft:
    
    {draft}
    
    Provide a detailed reflection on the strengths and weaknesses of this draft. 
    Focus on structure, tone, and whether the arguments are well-supported. 
    Suggest specific improvements for the final version.
    """

    response = CLIENT.chat.completions.create(
        model=model,
        messages=[{"role": "user", "content": prompt}],
        temperature=1.0,
    )

    return response.choices[0].message.content
```

```python
# test
unittests.test_reflect_on_draft(reflect_on_draft)
```

* **Step 3 – Revision:**

```python
def revise_draft(original_draft: str, reflection: str, model: str = "openai:gpt-4o") -> str:

    prompt = f"""
    You are a master editor. Your task is to rewrite the following essay draft based on a specific reflection/critique.
    
    ORIGINAL DRAFT:
    {original_draft}
    
    REFLECTION AND FEEDBACK:
    {reflection}
    
    INSTRUCTIONS:
    1. Incorporate all the suggestions from the reflection.
    2. Maintain a professional tone while improving clarity and flow.
    3. Ensure the final essay is a polished, ready-to-publish version.
    """

    response = CLIENT.chat.completions.create(
        model=model,
        messages=[{"role": "user", "content": prompt}],
        temperature=0.7,
    )

    return response.choices[0].message.content
```

```python
# test
unittests.test_revise_draft(revise_draft)
```

###  Test the Reflective Writing Workflow

```python
from utils import show_output

essay_prompt = "Should social media platforms be regulated by the government?"

# Agent 1 – Draft
draft = generate_draft(essay_prompt)

# Agent 2 – Reflection
feedback = reflect_on_draft(draft)

# Agent 3 – Revision
revised = revise_draft(draft, feedback)

show_output("Step 1 – Draft", draft, background="#fff8dc", text_color="#333333")
show_output("Step 2 – Reflection", feedback, background="#e0f7fa", text_color="#222222")
show_output("Step 3 – Revision", revised, background="#f3e5f5", text_color="#222222")
```
