## Image Generation

- **Generations**: Images from scratch, based on a text prompt
- **Edits**: Edited versions of images, where the model replaces areas of a pre-existing image based on a new prompt
- **Variations**: Variations of an existing image

### DALL-E 2 vs DALL-E 3

| | DALL-E 2 | DALL-E 3 |
|---|---|---|
| Sizes | 256x256, 512x512, 1024x1024 | 1024x1024, 1024x1792, 1792x1024 |
| Quality | standard | standard, hd |
| Max images (n) | up to 10 | 1 |

```python
from openai import OpenAI
from IPython import display

client = OpenAI()

response = client.images.generate(
    model="dall-e-2",
    prompt="a white siamese cat",
    size="1024x1024",
    n=1,
)

url = response.data[0].url
display.Image(url=url, width=512)
```
